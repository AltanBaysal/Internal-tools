Google Colab (Colaboratory), tarayıcınızda Python kodu yazıp çalıştırmanızı sağlayan **ücretsiz** bir araçtır.

## Hücreler (Cells) Nasıl Çalışır?

- Notebook **hücrelerden** oluşur. İki tür hücre vardır: **Kod** ve **Metin (Markdown)**.
- Bir hücreyi çalıştırmak için: hücreyi seçin ve **Shift + Enter** tuşlarına basın.
- Yeni hücre eklemek için üstteki **+ Code** veya **+ Text** butonlarını kullanın.
- Hücreleri sürükleyerek sırasını değiştirebilirsiniz.

# GPU (T4) Nasıl Açılır?

Google Colab size ücretsiz GPU verir. Açmak için şu adımları izleyin:

1. Üst menüden **Runtime** (Çalışma Zamanı) seçin
2. **Change runtime type** (Çalışma zamanı türünü değiştir) tıklayın
3. **Hardware accelerator** kısmından **T4 GPU** seçin
4. **Save** (Kaydet) butonuna basın

Bu kadar! Artık GPU kullanmaya hazırsınız.

```python
# GPU'nun bağlı olup olmadığını kontrol edelim

# nvidia-smi komutu GPU bilgilerini gösterir
!nvidia-smi

# PyTorch ile GPU'nun kullanılabilir olduğunu doğrulayalım
import torch

# GPU varsa True, yoksa False döner
print("GPU kullanılabilir mi?", torch.cuda.is_available())

# Bağlı GPU'nun adını yazdıralım
if torch.cuda.is_available():
    print("GPU adı:", torch.cuda.get_device_name(0))
```

# Şimdi Ne Yapacağız?

Basit bir örnek yapacağız:
1. İki sayı tanımlayacağız
2. Bunları toplayıp sonucu ekrana yazdıracağız
3. Sonucu bir dosyaya kaydedip bilgisayarımıza indireceğiz

Hadi başlayalım!

```python
# İki sayı tanımlayalım — bu değerleri istediğiniz gibi değiştirebilirsiniz!

# Birinci sayı
a = 10

# İkinci sayı
b = 20

# Tanımladığımız sayıları ekrana yazdıralım
print("a =", a)
print("b =", b)
```

```python
# İki sayıyı toplayalım
result = a + b

# Sonucu güzel bir formatta yazdıralım
print(f"{a} + {b} = {result}")
```

# Dosya İndirme

Colab'da oluşturduğunuz dosyaları bilgisayarınıza indirebilirsiniz.

Bunun için `google.colab.files` modülündeki `download()` fonksiyonunu kullanıyoruz.
Bu fonksiyon çağrıldığında, dosya otomatik olarak tarayıcınız üzerinden indirilir.

```python
# google.colab modülünden files fonksiyonunu içe aktaralım
from google.colab import files

# Sonucu bir dosyaya kaydedelim
# 'w' parametresi dosyayı yazma modunda açar
with open("result.txt", "w") as f:
    # Sonucu dosyaya yazalım
    f.write(f"Toplama Sonucu: {a} + {b} = {result}\n")

# Dosyanın başarıyla kaydedildiğini bildirelim
print("result.txt dosyası kaydedildi!")

# Dosyayı bilgisayarınıza indirin
files.download("result.txt")
```
