## 5. SmoothMix Checkpoint'leri (Civitai)

Version ID'leri dolduruldu:
- **I2V v2.0 High** → `2513182`
- **I2V v2.0 Low**  → `2513186`
- **T2V v3.0 High** → `2768924`
- **T2V v3.0 Low**  → `2768944`

Workflow `SmoothMix_I2V_v2_High.safetensors` gibi isimler bekliyor — indirme sonrası otomatik rename ediliyor.

Senin NSFW LoRA listen de altta — `loras/` klasörüne iniyor, Power LoRA Loader node'undan seçeceksin.

````python
# Version ID'ler
SMOOTHMIX_I2V_HIGH_V2 = 2513182
SMOOTHMIX_I2V_LOW_V2  = 2513186
SMOOTHMIX_T2V_HIGH_V3 = 2768924
SMOOTHMIX_T2V_LOW_V3  = 2768944

import os, glob, shutil, subprocess, struct, time

DIFF = f"{COMFY}/models/diffusion_models"
LORA = f"{COMFY}/models/loras"

# ============================================================
# Helpers (SRP)
# ============================================================

def log(msg, level="INFO"):
    icons = {"INFO":"ℹ️ ","OK":"✅","WARN":"⚠️ ","ERR":"❌"}
    print(f"{icons.get(level,'·')} [{time.strftime('%H:%M:%S')}] {msg}")

def human(b):
    for u in ['B','KB','MB','GB']:
        if b < 1024: return f"{b:.1f}{u}"
        b /= 1024
    return f"{b:.1f}TB"

def verify_safetensors(path):
    """Dosya gerçek safetensors mı? -> (ok: bool, msg: str)"""
    if not os.path.exists(path):
        return False, "yok"
    sz = os.path.getsize(path)
    if sz < 1_000_000:
        return False, f"çok küçük ({human(sz)})"
    with open(path, 'rb') as f:
        head = f.read(8)
    if head.startswith(b'<') or head.startswith(b'{"'):
        return False, "HTML/JSON error sayfası"
    try:
        jl = struct.unpack('<Q', head)[0]
        if 100 < jl < 200_000_000:
            return True, f"valid ({human(sz)})"
    except Exception:
        pass
    return False, "header bozuk"

def civitai_url(version_id):
    return f"https://civitai.com/api/download/models/{version_id}?token={CIVITAI_API_KEY}"

def cookie_header():
    return f"Cookie: __Secure-civitai-token={COOKIE_VALUE}"

# ============================================================
# Strategies — her biri tek sorumluluk: indir, valid mi dön
# Signature: (url, target_path) -> (ok: bool, elapsed: float)
# ============================================================

def download_with_aria2c(url, target_path):
    """Hızlı: 16 paralel bağlantı. Bazı Civitai URL'lerinde exit 22 verir."""
    t0 = time.time()
    target_dir, filename = os.path.split(target_path)
    cmd = [
        'aria2c', '-x', '16', '-s', '16', '-k', '1M',
        '--console-log-level=warn', '--summary-interval=30',
        '--header', cookie_header(),
        '--auto-file-renaming=false', '--allow-overwrite=true',
        '-d', target_dir, '-o', filename, url
    ]
    subprocess.run(cmd, capture_output=True, text=True, timeout=3600)
    ok, _ = verify_safetensors(target_path)
    return ok, time.time() - t0

def download_with_curl(url, target_path):
    """Güvenilir: tek bağlantı, redirect chain'i sağlam handle eder."""
    t0 = time.time()
    cmd = [
        'curl', '-sL', '-o', target_path,
        '-H', cookie_header(),
        '--max-time', '600', url
    ]
    subprocess.run(cmd, capture_output=True, text=True, timeout=650)
    ok, _ = verify_safetensors(target_path)
    return ok, time.time() - t0

# Strategy order: hızlıdan güvenilire
STRATEGIES = [
    ("aria2c", download_with_aria2c),
    ("curl",   download_with_curl),
]

# ============================================================
# Orchestrator
# ============================================================

def fetch_model(version_id, target_dir, filename, label):
    target = os.path.join(target_dir, filename)
    os.makedirs(target_dir, exist_ok=True)

    # 1. Already downloaded & valid?
    ok, msg = verify_safetensors(target)
    if ok:
        log(f"{label}: zaten var ({msg})")
        return True

    # 2. Try strategies in order
    url = civitai_url(version_id)
    for strat_name, strat_fn in STRATEGIES:
        # Önceki başarısız denemenin kalıntısını temizle
        if os.path.exists(target):
            os.remove(target)
        ok, elapsed = strat_fn(url, target)
        if ok:
            sz = os.path.getsize(target)
            log(f"{label}: ✓ {strat_name} ({elapsed:.0f}s, {human(sz)})", "OK")
            return True
        log(f"{label}: {strat_name} başarısız", "WARN")

    log(f"{label}: tüm strategy'ler başarısız", "ERR")
    return False

# ============================================================
# Çalıştır
# ============================================================

DOWNLOADS = [
    # (version_id, target_dir, filename, label)
    (SMOOTHMIX_I2V_HIGH_V2, DIFF, "SmoothMix_I2V_v2_High.safetensors",  "SmoothMix I2V v2 HIGH"),
    (SMOOTHMIX_I2V_LOW_V2,  DIFF, "SmoothMix_I2V_v2_Low.safetensors",   "SmoothMix I2V v2 LOW"),
    # (SMOOTHMIX_T2V_HIGH_V3, DIFF, "SmoothMix_T2V_High_v3.safetensors",  "SmoothMix T2V v3 HIGH"),    # T2V kullanılmıyor
    # (SMOOTHMIX_T2V_LOW_V3,  DIFF, "SmoothMix_T2V_Low_v3.safetensors",   "SmoothMix T2V v3 LOW"),     # T2V kullanılmıyor
    (2073605, LORA, "WAN_General_NSFW_HIGH.safetensors",  "WAN General NSFW HIGH"),
    (2083303, LORA, "WAN_General_NSFW_LOW.safetensors",   "WAN General NSFW LOW"),
    # (2209344, LORA, "Slop_Bounce_I2V_Low.safetensors",    "Slop Bounce LOW"),                # Şu an workflow'da yok
    # (2209354, LORA, "Slop_Bounce_I2V_High.safetensors",   "Slop Bounce HIGH"),               # Şu an workflow'da yok
    # (2553151, LORA, "DR34ML4Y_I2V_14B_HIGH_V2.safetensors", "DR34ML4Y HIGH"),                # Şu an workflow'da yok
    # (2553271, LORA, "DR34ML4Y_I2V_14B_LOW_V2.safetensors",  "DR34ML4Y LOW"),                 # Şu an workflow'da yok
]

results = [fetch_model(*d) for d in DOWNLOADS]

# ============================================================
# Özet + kritik dosya kontrolü
# ============================================================

print("\n" + "="*60)
log(f"ÖZET: {sum(results)}/{len(results)} başarılı",
    "OK" if all(results) else "WARN")

print("\n📂 diffusion_models/")
for f in sorted(glob.glob(f"{DIFF}/*.safetensors")):
    print(f"   {human(os.path.getsize(f))}  {os.path.basename(f)}")
print("\n📂 loras/")
for f in sorted(glob.glob(f"{LORA}/*.safetensors")):
    print(f"   {human(os.path.getsize(f))}  {os.path.basename(f)}")

REQUIRED = [
    f"{DIFF}/SmoothMix_I2V_v2_High.safetensors",
    f"{DIFF}/SmoothMix_I2V_v2_Low.safetensors",
    f"{DIFF}/SmoothMix_T2V_High_v3.safetensors",
    f"{DIFF}/SmoothMix_T2V_Low_v3.safetensors",
]
missing = [os.path.basename(p) for p in REQUIRED if not os.path.exists(p)]
print()
if missing:
    log(f"EKSİK kritik dosyalar: {missing}", "ERR")
else:
    log("Tüm SmoothMix checkpoint'leri yerinde", "OK")
````

## 7. API Modülleri

Her fonksiyon tek sorumluluk. Bir kere çalıştır, sonra değiştirmesen yeter.

````python
import json, copy, time, uuid, random, requests
from pathlib import Path

# =============================================================
# Template I/O — SRP
# =============================================================
def load_workflow(path):
    with open(path) as f: return json.load(f)

def clone(workflow):
    return copy.deepcopy(workflow)

# =============================================================
# Patchers — her biri tek alana enjeksiyon (SRP)
# =============================================================
POSITIVE_PROMPT_NODE = "333:291"
START_IMAGE_NODE     = "338"
END_IMAGE_NODE       = "342"
SEED_NODE            = "327"
LORA_LOADER_NODES    = ["324", "325"]

def set_prompt(workflow, prompt):
    workflow[POSITIVE_PROMPT_NODE]["inputs"]["prompt"] = prompt

def set_images(workflow, start_filename, end_filename):
    workflow[START_IMAGE_NODE]["inputs"]["image"] = start_filename
    workflow[END_IMAGE_NODE]["inputs"]["image"]   = end_filename

def set_seed(workflow, seed):
    workflow[SEED_NODE]["inputs"]["seed"] = seed

def set_lora_strengths(workflow, strength_map):
    for node_id in LORA_LOADER_NODES:
        node = workflow.get(node_id)
        if not node: continue
        for v in node["inputs"].values():
            if isinstance(v, dict) and "lora" in v:
                if v["lora"] in strength_map:
                    v["strength"] = strength_map[v["lora"]]

def apply_config(workflow, start_filename, end_filename, seed):
    set_prompt(workflow, POSITIVE_PROMPT)
    set_images(workflow, start_filename, end_filename)
    set_seed(workflow, seed)
    set_lora_strengths(workflow, LORA_STRENGTHS)
    return workflow

# =============================================================
# ComfyUI HTTP client
# =============================================================
class ComfyClient:
    def __init__(self, base_url):
        self.base = base_url.rstrip("/")
        self.client_id = str(uuid.uuid4())

    def upload_image(self, local_path):
        name = Path(local_path).name
        with open(local_path, "rb") as f:
            r = requests.post(
                f"{self.base}/upload/image",
                files={"image": (name, f)},
                data={"overwrite": "true"},
                timeout=60,
            )
        r.raise_for_status()
        return r.json()["name"]

    def submit(self, workflow):
        r = requests.post(
            f"{self.base}/prompt",
            json={"prompt": workflow, "client_id": self.client_id},
            timeout=30,
        )
        r.raise_for_status()
        data = r.json()
        if data.get("node_errors"):
            raise RuntimeError(f"Workflow hatası: {data['node_errors']}")
        return data["prompt_id"]

    def wait(self, prompt_id, poll_interval=2.0):
        last = ""
        while True:
            history = requests.get(f"{self.base}/history/{prompt_id}", timeout=10).json()
            if prompt_id in history:
                entry = history[prompt_id]
                status = entry.get("status", {})
                if status.get("status_str") == "error":
                    raise RuntimeError(f"ComfyUI error: {status.get('messages')}")
                return entry
            q = requests.get(f"{self.base}/queue", timeout=10).json()
            msg = f"⏳ running:{len(q.get('queue_running',[]))} pending:{len(q.get('queue_pending',[]))}"
            if msg != last:
                print(msg); last = msg
            time.sleep(poll_interval)

    def download_video(self, history_entry, save_path):
        for node_output in history_entry["outputs"].values():
            for key in ("gifs", "videos", "images"):
                for item in node_output.get(key, []):
                    if item.get("filename", "").lower().endswith((".mp4", ".webm", ".mov")):
                        params = {
                            "filename":  item["filename"],
                            "subfolder": item.get("subfolder", ""),
                            "type":      item.get("type", "output"),
                        }
                        r = requests.get(f"{self.base}/view", params=params, timeout=120)
                        r.raise_for_status()
                        with open(save_path, "wb") as f:
                            f.write(r.content)
                        return save_path
        raise RuntimeError("Video çıktısı bulunamadı")

# =============================================================
# Orchestrator — batch mode + timing
# =============================================================
def resolve_seed(index):
    if SEED_START is None:
        return random.randint(0, 2**31 - 1)
    return SEED_START + index

def output_path_for(input_image_path, render_index, seed):
    """input adıyla eşleşen output path üret.
    NUM_RENDERS=1 → photo.png → photo.mp4
    NUM_RENDERS>1 → photo.png → photo_seed12345.mp4
    """
    stem = Path(input_image_path).stem
    if NUM_RENDERS == 1:
        return f"{OUTPUT_DIR}/{stem}.mp4"
    return f"{OUTPUT_DIR}/{stem}_seed{seed}.mp4"

def render_one(client, template, image_remote_name, image_local_path, render_index):
    t0 = time.time()
    workflow = clone(template)
    seed = resolve_seed(render_index)
    apply_config(workflow, image_remote_name, image_remote_name, seed)
    save_path = output_path_for(image_local_path, render_index, seed)
    print(f"   → Render {render_index+1}/{NUM_RENDERS} (seed={seed})")
    pid = client.submit(workflow)
    history = client.wait(pid)
    client.download_video(history, save_path)
    elapsed = time.time() - t0
    print(f"   ✓ ({elapsed:.1f}s) → {Path(save_path).name}")
    return save_path, elapsed

def render_image(client, template, image_local_path, image_index, total_images):
    """Tek bir görsel için NUM_RENDERS kadar render yap."""
    name = Path(image_local_path).name
    print(f"\n📸 Görsel {image_index+1}/{total_images}: {name}")
    image_remote = client.upload_image(image_local_path)
    results = []
    for i in range(NUM_RENDERS):
        p, e = render_one(client, template, image_remote, image_local_path, i)
        results.append((p, e))
    return results

def run():
    client = ComfyClient(COMFYUI_URL)
    template = load_workflow(WORKFLOW_PATH)
    total_t0 = time.time()
    all_results = []

    for idx, img_path in enumerate(INPUT_IMAGES):
        results = render_image(client, template, img_path, idx, len(INPUT_IMAGES))
        all_results.extend(results)

    total = time.time() - total_t0
    paths = [r[0] for r in all_results]
    elapseds = [r[1] for r in all_results]
    avg = sum(elapseds) / len(elapseds) if elapseds else 0
    print(f"\n{'='*60}")
    print(f"🎉 {len(INPUT_IMAGES)} görsel × {NUM_RENDERS} render = {len(paths)} video")
    print(f"   Toplam: {total:.1f}s | Ortalama: {avg:.1f}s/video")
    return paths

print("✓ Modüller yüklendi (batch mode)")
````
