## 4. SmoothMix Checkpoint'leri (Civitai)

Version ID'leri dolduruldu:
- **I2V v2.0 High** → `2513182`
- **I2V v2.0 Low**  → `2513186`
- **T2V v3.0 High** → `2768924`
- **T2V v3.0 Low**  → `2768944`

Workflow `SmoothMix_I2V_v2_High.safetensors` gibi isimler bekliyor — indirme sonrası otomatik rename ediliyor.

Senin NSFW LoRA listen de altta — `loras/` klasörüne iniyor, Power LoRA Loader node'undan seçeceksin.

```python
# Version ID'ler
SMOOTHMIX_I2V_HIGH_V2 = 2513182
SMOOTHMIX_I2V_LOW_V2  = 2513186
SMOOTHMIX_T2V_HIGH_V3 = 2768924
SMOOTHMIX_T2V_LOW_V3  = 2768944

import os, glob, shutil, subprocess, struct, time

def log(msg, level="INFO"):
    icons = {"INFO":"ℹ️","OK":"✅","WARN":"⚠️","ERR":"❌"}
    print(f"{icons.get(level,'·')} [{time.strftime('%H:%M:%S')}] {msg}")

def human(b):
    for u in ['B','KB','MB','GB']:
        if b < 1024: return f"{b:.1f}{u}"
        b /= 1024
    return f"{b:.1f}TB"

def verify_safetensors(path):
    """Dosya gerçek safetensors mı? HTML/error sayfası filtresi."""
    if not os.path.exists(path): return False, "yok"
    sz = os.path.getsize(path)
    if sz < 10_000_000: return False, f"çok küçük ({human(sz)})"
    with open(path, 'rb') as f:
        head = f.read(8)
    if head.startswith(b'<') or head.startswith(b'{"'):
        return False, "HTML/JSON error"
    try:
        jl = struct.unpack('<Q', head)[0]
        if 100 < jl < 200_000_000:
            return True, f"valid ({human(sz)})"
    except Exception:
        pass
    return False, "header bozuk"

DIFF = f"{COMFY}/models/diffusion_models"
LORA = f"{COMFY}/models/loras"

def civitai_download(version_id, target_dir, target_filename, label):
    """Cookie + token ile aria2c indirme. Sonra safetensors doğrulama."""
    target = f"{target_dir}/{target_filename}"

    # Zaten var ve sağlamsa atla
    if os.path.exists(target):
        ok, msg = verify_safetensors(target)
        if ok:
            log(f"{label}: zaten var ({msg})")
            return True
        else:
            log(f"{label}: var ama bozuk ({msg}), yeniden iniyor", "WARN")
            os.remove(target)

    log(f"{label} (v{version_id}): indiriliyor")
    t0 = time.time()
    os.makedirs(target_dir, exist_ok=True)
    before = set(glob.glob(f"{target_dir}/*"))

    url = f"https://civitai.com/api/download/models/{version_id}?token={CIVITAI_API_KEY}"
    cmd = [
        'aria2c', '-x', '16', '-s', '16', '-k', '1M',
        '--console-log-level=warn', '--summary-interval=30',
        '--header', f'Cookie: __Secure-civitai-token={COOKIE_VALUE}',
        '--content-disposition-default-utf8=true',
        '--auto-file-renaming=false', '--allow-overwrite=true',
        '-d', target_dir,
        '-o', target_filename,
        url
    ]
    r = subprocess.run(cmd, capture_output=True, text=True, timeout=3600)

    # Civitai bazen Content-Disposition'a göre kendi ismini koyabiliyor — düzelt
    after = set(glob.glob(f"{target_dir}/*"))
    new = [f for f in (after - before) if not f.endswith('.aria2')]
    if new and not os.path.exists(target):
        for nf in new:
            shutil.move(nf, target)

    elapsed = time.time() - t0
    ok, msg = verify_safetensors(target)
    if ok:
        log(f"{label}: ✓ {msg} ({elapsed:.0f}s)", "OK")
        return True
    else:
        log(f"{label}: ✗ {msg}", "ERR")
        if r.stdout:
            for line in r.stdout.strip().split('\n')[-3:]:
                print(f"     {line}")
        return False

# --- 4 SmoothMix Checkpoint ---
results = []
results.append(civitai_download(SMOOTHMIX_I2V_HIGH_V2, DIFF, "SmoothMix_I2V_v2_High.safetensors",  "SmoothMix I2V v2 HIGH"))
results.append(civitai_download(SMOOTHMIX_I2V_LOW_V2,  DIFF, "SmoothMix_I2V_v2_Low.safetensors",   "SmoothMix I2V v2 LOW"))
results.append(civitai_download(SMOOTHMIX_T2V_HIGH_V3, DIFF, "SmoothMix_T2V_High_v3.safetensors",  "SmoothMix T2V v3 HIGH"))
results.append(civitai_download(SMOOTHMIX_T2V_LOW_V3,  DIFF, "SmoothMix_T2V_Low_v3.safetensors",   "SmoothMix T2V v3 LOW"))

# --- NSFW LoRA'lar ---
EXTRA_LORAS = [
    (2073605, "WAN_General_NSFW_HIGH.safetensors", "WAN General NSFW HIGH"),
    (2083303, "WAN_General_NSFW_LOW.safetensors",  "WAN General NSFW LOW"),
    (2209344, "Slop_Bounce_I2V_Low.safetensors",   "Slop Bounce I2V Low"),
    (2209354, "Slop_Bounce_I2V_High.safetensors",  "Slop Bounce I2V High"),
    (2553151, "DR34ML4Y_I2V_14B_HIGH_V2.safetensors", "DR34ML4Y HIGH"),
    (2553271, "DR34ML4Y_I2V_14B_LOW_V2.safetensors",  "DR34ML4Y LOW"),
]
for vid, fname, lbl in EXTRA_LORAS:
    results.append(civitai_download(vid, LORA, fname, lbl))

# --- Özet ---
print("\n" + "="*60)
ok_count = sum(results)
total = len(results)
log(f"ÖZET: {ok_count}/{total} başarılı", "OK" if ok_count == total else "WARN")

print("\n📂 diffusion_models/")
for f in sorted(glob.glob(f"{DIFF}/*.safetensors")):
    print(f"   {human(os.path.getsize(f))}  {os.path.basename(f)}")
print("\n📂 loras/")
for f in sorted(glob.glob(f"{LORA}/*.safetensors")):
    print(f"   {human(os.path.getsize(f))}  {os.path.basename(f)}")

# Workflow'un kritik dosyalarını son kontrol
expected = [
    f"{DIFF}/SmoothMix_I2V_v2_High.safetensors",
    f"{DIFF}/SmoothMix_I2V_v2_Low.safetensors",
    f"{DIFF}/SmoothMix_T2V_High_v3.safetensors",
    f"{DIFF}/SmoothMix_T2V_Low_v3.safetensors",
]
missing = [e for e in expected if not os.path.exists(e)]
print()
if missing:
    log(f"EKSİK SmoothMix: {[os.path.basename(m) for m in missing]}", "ERR")
else:
    log("Tüm SmoothMix checkpoint'leri yerinde", "OK")
```

## 5. ComfyUI'yi Başlat + Cloudflared Tunnel

Bu hücre sürekli çalışır. Loglarda `https://xxx.trycloudflare.com` URL'sini göreceksin → tıkla.

**ComfyUI açıldıktan sonra:**
1. Üst menüden **Manager** → **Install Missing Custom Nodes** (workflow yüklendikten sonra eksik kalan varsa otomatik bulur)
2. `WAN_2_2_Smooth_Workflow_v5_0.json` dosyasını arayüze sürükle-bırak
3. Workflow'un içinden hangi modu kullanacağını seç (I2V / T2V / First2Last / Audio)
4. UNETLoader node'larında dropdown'ların doğru SmoothMix dosyalarını gösterdiğini kontrol et
5. **HIGH LORA LOADER** ve **LOW LORA LOADER** node'larından istediğin NSFW LoRA'ları aktif et

```python
import subprocess, threading, time, socket, re

def tunnel_thread(port):
    while True:
        time.sleep(1)
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        if s.connect_ex(('127.0.0.1', port)) == 0:
            s.close()
            break
        s.close()
    print("\n🌐 ComfyUI hazır, cloudflared tunnel açılıyor...\n")
    p = subprocess.Popen(
        ["cloudflared", "tunnel", "--url", f"http://127.0.0.1:{port}"],
        stdout=subprocess.PIPE, stderr=subprocess.PIPE
    )
    for line in p.stderr:
        l = line.decode(errors='ignore')
        m = re.search(r'https://[\w\-]+\.trycloudflare\.com', l)
        if m:
            print("\n" + "="*70)
            print(f"  🚀 ComfyUI URL: {m.group(0)}")
            print("="*70 + "\n")
            break
    p.wait()

threading.Thread(target=tunnel_thread, daemon=True, args=(8188,)).start()

%cd /content/ComfyUI
!python main.py --listen 0.0.0.0 --port 8188
```
